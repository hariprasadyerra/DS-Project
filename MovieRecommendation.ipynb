```python
import numpy as np
from scipy.sparse import csr_matrix

# Pivot ratings data to create a user-item matrix
ratings_data = ratings_small.pivot(index='userId', columns='movieId', values='rating').fillna(0)
user_movie_matrix = csr_matrix(ratings_data.values)
```

```python
from sklearn.neighbors import NearestNeighbors

# Train the User-Based Collaborative Filtering Model using Nearest Neighbors
user_knn = NearestNeighbors(metric='cosine', algorithm='brute')
user_knn.fit(user_movie_matrix)
```

```python
def get_user_recommendations(user_id, k=10):
    # Find the user index based on user_id
    user_index = ratings_small.index[ratings_small['userId'] == user_id].tolist()
    if not user_index:
        return []  # Return empty list if user not found

    user_index = user_index[0]  # Get the first index (if there are multiple)

    # Find the k-nearest neighbors of the user
    distances, indices = user_knn.kneighbors(user_movie_matrix[user_index, :].reshape(1, -1), n_neighbors=k+1)
    similar_users = indices.flatten()

    # List to store the recommended movie IDs
    recommended_movies = []

    # Iterate through the similar users
    for similar_user in similar_users:
        # Get the ratings for the similar user from the ratings_small DataFrame
        similar_user_ratings = ratings_small[ratings_small['userId'] == similar_user]

        # Sort the similar user's ratings by the rating value, in descending order
        top_movies = similar_user_ratings.sort_values(by='rating', ascending=False)

        # Get the top 5 recommended movies (by movieId)
        recommended_movies.extend(top_movies.head(5)['movieId'].values)

    # Return top unique movie recommendations
    return list(set(recommended_movies[:k]))

# Example usage
user_id = 1  # Replace with a specific user ID
print(f"Recommended movies for User {user_id} (User-Based CF):")
user_response = get_user_recommendations(user_id)
for idx, movie in enumerate(user_response, 1):
    movie_title = set(movies_metadata.loc[movies_metadata['id'] == str(movie), 'title'].values)
    rating_movie_title = movies_metadata.loc[movies_metadata['movieId'] == float(movie), 'title'].values
    movie_title = "".join(movie_title)
    print(f"{idx}. {movie_title if len(movie_title)>0 else rating_movie_title if len(rating_movie_title)>0 else movie}")
```

```python
# Train the Item-Based Collaborative Filtering Model
movie_knn = NearestNeighbors(metric='cosine', algorithm='brute')
movie_knn.fit(user_movie_matrix.T)  # Transpose for item-item similarity
```

```python
# Function for Item-Based Recommendations
def get_item_recommendations(movie_id, k=10):
    movie_index = ratings_data.columns.get_loc(movie_id)
    distances, indices = movie_knn.kneighbors(user_movie_matrix.T[movie_index, :].reshape(1, -1), n_neighbors=k+1)
    similar_movies = indices.flatten()

    return list(ratings_data.columns[similar_movies])

# Example usage
movie_id = 31  # Replace with a specific movie ID
item_response = get_item_recommendations(movie_id)
print(f"Recommended movies similar to Movie {movie_id} (Item-Based CF):")
for idx, movie in enumerate(item_response, 1):
    movie_title = set(movies_metadata.loc[movies_metadata['id'] == str(movie), 'title'].values)
    rating_movie_title = movies_metadata.loc[movies_metadata['movieId'] == float(movie), 'title'].values
    movie_title = "".join(movie_title)
    print(f"{idx}. {movie_title if len(movie_title)>0 else rating_movie_title if len(rating_movie_title)>0 else movie}")
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

# RMSE Calculation Function
def rmse(pred, actual):
    return sqrt(mean_squared_error(pred, actual))

# Example RMSE Calculation for User-Based Collaborative Filtering
test_data = ratings_small.sample(frac=0.2, random_state=42)  # Use 20% of data for testing
predictions = []
actual = []

for index, row in test_data.iterrows():
    user_id, movie_id, actual_rating = int(row['userId']), int(row['movieId']), row['rating']
    try:
        pred = np.mean([ratings_data.iloc[user_id, movie] for movie in get_user_recommendations(user_id)])
        predictions.append(pred)
        actual.append(actual_rating)
    except:
        continue

print("User-Based CF RMSE:", rmse(predictions, actual))
```

```python
# Convert genres and cast into strings and fill NaN values with empty strings
movies_metadata['genres'] = movies_metadata['genres'].fillna('').astype(str)
movies_metadata['cast'] = movies_metadata['cast'].fillna('').astype(str)

# Combine genres and cast into one string
movies_metadata['combined_features'] = movies_metadata['genres'] + " " + movies_metadata['cast']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

# Initialize TF-IDF Vectorizer
tfidf = TfidfVectorizer(stop_words='english')

# Transform combined features into TF-IDF matrix
movies_metadata = movies_metadata.sample(n=5000, random_state=42)
tfidf_matrix = tfidf.fit_transform(movies_metadata['combined_features'])
```

```python
movies_metadata["title"].tail()
```

```python
# Compute the cosine similarity matrix based on the TF-IDF matrix
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
# Function to get movie recommendations based on content similarity
def get_content_based_recommendations(title, top_n=10, cosine_sim=cosine_sim):
    indices = pd.Series(movies_metadata.index, index=movies_metadata['title']).drop_duplicates()

    # Get the index of the movie that matches the title
    idx = indices.get(title, None)
    if idx is None:
        return "Movie title not found in the dataset."

    # Retrieve similarity scores for all movies with the input movie
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort the movies by similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Select top N most similar movies
    sim_scores = sim_scores[1:top_n+1]

    # Get the movie indices
    movie_indices = [i[0] for i in sim_scores]

    # Return the top N similar movie titles
    data = list(set(movies_metadata['id'].iloc[movie_indices]))
    final_data = []
    for x in data:
      final_data.append(int(x))

    return final_data
```

```python
movie_title = "Sunset Boulevard"  # Replace with a valid title in your dataset
print(f"Movies similar to '{movie_title}' (Content-Based):")
content_response = get_content_based_recommendations(movie_title)
for idx, movie in enumerate(content_response, 1):
    movie_title = set(movies_metadata.loc[movies_metadata['id'] == str(movie), 'title'].values)
    movie_title = "".join(movie_title)
    print(f"{idx}: {movie_title}")
```

```python
def hybrid_recommendations(user_id, movie_title, n_recommendations=10):
    # Content-based recommendations
    content_based_recs = get_content_based_recommendations(movie_title)

    # Collaborative filtering recommendations
    collaborative_recs = get_user_recommendations(user_id)

    # Combine both recommendations and remove duplicates
    hybrid_recs = list(set(content_based_recs).union(set(collaborative_recs)))
    return hybrid_recs[:n_recommendations]
```

```python
user_id = 1
movie_title = "Sunset Boulevard"

recommendations = hybrid_recommendations(user_id, movie_title)
print("Hybrid Recommendations:")
for idx, movie in enumerate(recommendations, 1):
    movie_title = set(movies_metadata.loc[movies_metadata['id'] == str(movie), 'title'].values)
    rating_movie_title = movies_metadata.loc[movies_metadata['movieId'] == float(movie), 'title'].values
    movie_title = "".join(movie_title)
    print(f"{idx}. {movie_title if len(movie_title)>0 else rating_movie_title if len(rating_movie_title)>0 else movie}")
```

**Metrics & comparision of all three models**

```python

def content_precision_recall(ground_truth, recommendations, k=10):
    precision_list = []
    recall_list = []

    for user_id in ground_truth:
        # Get the top-k recommended items for this user
        recommended_items = recommendations(user_id)[:k]
        actual_items = ground_truth[user_id]
        # Calculate Precision
        relevant_recommended_items = set(recommended_items).intersection(set(actual_items))
        precision = len(relevant_recommended_items) / len(recommended_items) if len(recommended_items) > 0 else 0
        precision_list.append(precision)

        # Calculate Recall
        recall = len(relevant_recommended_items) / len(actual_items) if len(actual_items) > 0 else 0
        recall_list.append(recall)

    # Calculate average precision and recall across all users
    avg_precision = sum(precision_list) / len(precision_list)
    avg_recall = sum(recall_list) / len(recall_list)

    return avg_precision, avg_recall
```

```python

def collabrative_precision_recall(ground_truth, recommendations, k=10):
    precision_list = []
    recall_list = []

    for user_id in ground_truth:
        # Get the top-k recommended items for this user
        recommended_items = recommendations("Sunset Boulevard")[:k]
        actual_items = ground_truth[user_id]
        # Calculate Precision
        relevant_recommended_items = set(recommended_items).intersection(set(actual_items))
        precision = len(relevant_recommended_items) / len(recommended_items) if len(recommended_items) > 0 else 0
        precision_list.append(precision)

        # Calculate Recall
        recall = len(relevant_recommended_items) / len(actual_items) if len(actual_items) > 0 else 0
        recall_list.append(recall)

    # Calculate average precision and recall across all users
    avg_precision = sum(precision_list) / len(precision_list)
    avg_recall = sum(recall_list) / len(recall_list)

    return avg_precision, avg_recall
```

```python

def hybrid_precision_recall(ground_truth, recommendations, k=10):
    precision_list = []
    recall_list = []

    for user_id in ground_truth:
        # Get the top-k recommended items for this user
        recommended_items = recommendations(user_id, "Sunset Boulevard")[:k]
        actual_items = ground_truth[user_id]
        # Calculate Precision
        relevant_recommended_items = set(recommended_items).intersection(set(actual_items))
        precision = len(relevant_recommended_items) / len(recommended_items) if len(recommended_items) > 0 else 0
        precision_list.append(precision)

        # Calculate Recall
        recall = len(relevant_recommended_items) / len(actual_items) if len(actual_items) > 0 else 0
        recall_list.append(recall)

    # Calculate average precision and recall across all users
    avg_precision = sum(precision_list) / len(precision_list)
    avg_recall = sum(recall_list) / len(recall_list)

    return avg_precision, avg_recall
```
