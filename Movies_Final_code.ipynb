## Cast

We have cast_id and character for each cast in our dataset as we are most interested in the character of cast not on its id, we only keep its name by removing the extra unnecessary information

```python
def cast_names(data):
    actors = []
    counter = 0
    for i in ast.literal_eval(data):
        if counter != 3:
            actors.append(i['name'])
            counter += 1
    return actors
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.neighbors import NearestNeighbors
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from scipy.sparse import csr_matrix
from math import sqrt
```

```python
# Assuming df is your DataFrame
# Extract relevant columns: userId, movieId, rating
ratings = df[['userId', 'movieId', 'rating']].copy()

# Create a lookup dictionary for movie titles
movie_title_lookup = df.set_index('movieId')['title'].to_dict()

# Create pivot table for user-item interaction
user_item_matrix = ratings.pivot_table(index='userId', columns='movieId', values='rating').fillna(0)

# Convert the matrix to CSR for efficient computation
user_item_matrix_csr = csr_matrix(user_item_matrix.values)
```

```python
# Fit KNN model
model_knn_user = NearestNeighbors(metric='cosine', algorithm='brute')
model_knn_user.fit(user_item_matrix_csr)
```

```python
def recommend_movies_user(user_id, n_recommendations=5):
    # Find neighbors
    distances, indices = model_knn_user.kneighbors(
        user_item_matrix.loc[user_id, :].values.reshape(1, -1),
        n_neighbors=n_recommendations + 1
    )

    print(f"Recommendations for User {user_id}:\n")
    for i in range(1, len(distances.flatten())):
        movie_id = user_item_matrix.columns[indices.flatten()[i]]
        movie_title = movie_title_lookup.get(movie_id, "Unknown Title")
        distance = distances.flatten()[i]
        print(f"{i}: Movie ID {movie_id} ({movie_title}), with distance of {distance:.4f}:")


```

```python
def calculate_rmse(actual, predicted):
    actual_values = actual.values.flatten()
    predicted_values = predicted.values.flatten()
    mask = actual_values > 0  # Consider only non-zero ratings
    return sqrt(mean_squared_error(actual_values[mask], predicted_values[mask]))
```

```python
# Evaluate User-Based Filtering
user_distances, user_indices = model_knn_user.kneighbors(user_item_matrix.values)
user_predictions = np.mean(user_item_matrix.values[user_indices], axis=1)
rmse_user = calculate_rmse(user_item_matrix, pd.DataFrame(user_predictions))
print(f"User-Based Collaborative Filtering RMSE: {rmse_user:.4f}")
```

```python
# Transpose user-item matrix for item-based filtering
item_item_matrix = user_item_matrix.T
item_item_matrix_csr = csr_matrix(item_item_matrix.values)

# Fit KNN model
model_knn_item = NearestNeighbors(metric='cosine', algorithm='brute')
model_knn_item.fit(item_item_matrix_csr)
```

```python
def recommend_movies_item(movie_id, n_recommendations=5):
    # Find neighbors
    distances, indices = model_knn_item.kneighbors(
        item_item_matrix.loc[movie_id, :].values.reshape(1, -1),
        n_neighbors=n_recommendations + 1
    )

    print(f"Recommendations for Movie ID {movie_id}:\n")
    for i in range(1, len(distances.flatten())):
        similar_movie_id = item_item_matrix.index[indices.flatten()[i]]
        similar_movie_title = movie_title_lookup.get(similar_movie_id, "Unknown Title")
        distance = distances.flatten()[i]
        print(f"{i}: Movie ID {similar_movie_id} ({similar_movie_title}), with distance of {distance:.4f}:")
```

```python

# Evaluate Item-Based Filtering
item_distances, item_indices = model_knn_item.kneighbors(item_item_matrix.values)
item_predictions = np.mean(item_item_matrix.values[item_indices], axis=1)
rmse_item = calculate_rmse(item_item_matrix, pd.DataFrame(item_predictions))
print(f"Item-Based Collaborative Filtering RMSE: {rmse_item:.4f}")
```

```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import normalize
import pandas as pd

# Fill missing values and combine features
df['genres'] = df['genres'].fillna('').astype(str)
df['cast'] = df['cast'].fillna('').astype(str)
df['title'] = df['title'].fillna('Unknown')
df['keywords'] = df['keywords'].fillna('').astype(str)
df['overview'] = df['overview'].fillna('').astype(str)

# Updated combined features
df['combined_features'] = (
    df['genres'] + " " + df['cast'] + " " + df['keywords'] + " " + df['overview']
)

```

```python
# Initialize TF-IDF Vectorizer with n-grams
tfidf = TfidfVectorizer(stop_words='english', ngram_range=(1, 2), max_features=10000)

# Sample and reset the dataframe to ensure indices align
sample_size = min(20000, len(df))
movies_metadata = df.sample(n=sample_size, random_state=42).reset_index(drop=True)

# Remove duplicates based on combined_features
movies_metadata = movies_metadata.drop_duplicates(subset='combined_features').reset_index(drop=True)

# Fit and transform the TF-IDF matrix
tfidf_matrix = tfidf.fit_transform(movies_metadata['combined_features'])

# Normalize the TF-IDF matrix
tfidf_matrix = normalize(tfidf_matrix)

```

```python
# Compute cosine similarity
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Debug: Check cosine similarity matrix
np.fill_diagonal(cosine_sim, 0)  # Set diagonal to 0 to ignore self-similarity
print(f"Max similarity (excluding diagonal): {np.max(cosine_sim)}")
print(f"Number of similarities > 0.99: {np.sum(cosine_sim > 0.99)}")

# Map movie titles to new indices
indices = pd.Series(movies_metadata.index, index=movies_metadata['title']).drop_duplicates()

```

```python
# Content-Based Recommendation Function
def get_content_based_recommendations(title, top_n=10, cosine_sim=cosine_sim):
    idx = indices.get(title, None)

    if idx is None:
        return [(f"Movie '{title}' not found in the dataset.", 0.0)]

    # Calculate similarity scores for all movies
    sim_scores = list(enumerate(cosine_sim[idx].flatten()))

    # Filter out NaN values and the movie itself
    sim_scores = [(i, score) for i, score in sim_scores if not np.isnan(score) and i != idx]

    # Sort by similarity score in descending order
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get indices of the most similar movies
    movie_indices = [i[0] for i in sim_scores[:top_n]]

    # Retrieve movie titles and similarity scores
    recommended_movies = [(movies_metadata.loc[i, 'title'], sim_scores[j][1]) for j, i in enumerate(movie_indices)]

    return recommended_movies


    # Sort and return top recommendations
    sorted_recommendations = sorted(combined_scores.items(), key=lambda x: x[1], reverse=True)
    final_recommendations = [(rec[0], rec[1]) for rec in sorted_recommendations[:n_recommendations]]

    return final_recommendations
```

```python
def hybrid_recommendations(user_id, movie_title, n_recommendations=10, collaborative_weight=0.5, content_weight=0.5):
    """
    Generate hybrid recommendations combining collaborative (user-based) and content-based approaches.

    Parameters:
        user_id (int): User ID for collaborative filtering.
        movie_title (str): Movie title for content-based filtering.
        n_recommendations (int): Number of recommendations to generate.
        collaborative_weight (float): Weight for collaborative scores.
        content_weight (float): Weight for content-based scores.

    Returns:
        List[Tuple[str, float]]: List of recommended movies with hybrid scores.
    """
    # Collaborative recommendations
    collaborative_recs = []
    try:
        distances, indices = model_knn_user.kneighbors(
            user_item_matrix.loc[user_id, :].values.reshape(1, -1),
            n_neighbors=n_recommendations + 1
        )
        for i in range(1, len(distances.flatten())):
            movie_id = user_item_matrix.columns[indices.flatten()[i]]
            movie_title_collab = movie_title_lookup.get(movie_id, "Unknown Title")
            collaborative_recs.append((movie_title_collab, 1 - distances.flatten()[i]))
    except KeyError:
        print(f"User ID {user_id} not found in collaborative filtering data.")

    # Content-based recommendations
    content_recs = get_content_based_recommendations(movie_title, top_n=50)

    # Normalize scores for combination
    max_collab_score = max((score for _, score in collaborative_recs), default=1)
    max_content_score = max((score for _, score in content_recs), default=1)

    collaborative_recs = [(movie, score / max_collab_score) for movie, score in collaborative_recs]
    content_recs = [(movie, score / max_content_score) for movie, score in content_recs]

    # Combine scores
    combined_scores = {}
    for movie, score in collaborative_recs:
        combined_scores[movie] = combined_scores.get(movie, 0) + collaborative_weight * score

    for movie, score in content_recs:
        combined_scores[movie] = combined_scores.get(movie, 0) + content_weight * score

    # Sort and return top recommendations
    sorted_recommendations = sorted(combined_scores.items(), key=lambda x: x[1], reverse=True)
    final_recommendations = [(rec[0], rec[1]) for rec in sorted_recommendations[:n_recommendations]]

    return final_recommendations


```

```python
# Precision and Recall Calculation Function
def calculate_precision_recall(ground_truth, recommendations, k=10):
    """
    Calculate precision and recall for recommendations.
    """
    precision_list = []
    recall_list = []

    for user_id, actual_items in ground_truth.items():
        recommended_items = recommendations(user_id)[:k]  # Top-k recommendations

        relevant_recommended_items = set(recommended_items).intersection(set(actual_items))

        precision = len(relevant_recommended_items) / len(recommended_items) if recommended_items else 0
        recall = len(relevant_recommended_items) / len(actual_items) if actual_items else 0

        precision_list.append(precision)
        recall_list.append(recall)

    avg_precision = sum(precision_list) / len(precision_list) if precision_list else 0
    avg_recall = sum(recall_list) / len(recall_list) if recall_list else 0

    return avg_precision, avg_recall

# Ground Truth for Testing
user_ground_truth = {
    1: user_based_recommendations(1, 100),
    2: user_based_recommendations(2, 100)
}
item_ground_truth = {
    1: item_based_recommendations(31, 100),
    2: item_based_recommendations(31, 100)
}
content_ground_truth = {
    1: content_based_recommendations("Sunset Boulevard", 100),
    2: content_based_recommendations("Sunset Boulevard", 100)
}
hybrid_ground_truth = {
    1: hybrid_recommendations(1, "Sunset Boulevard", 100),
    2: hybrid_recommendations(2, "Sunset Boulevard", 100)
}

# User-Based Collaborative Filtering
user_precision, user_recall = calculate_precision_recall(user_ground_truth, user_based_recommendations)
print(f"User-Based Collaborative Filtering - Precision: {user_precision}, Recall: {user_recall}")

# Item-Based Collaborative Filtering
item_precision, item_recall = calculate_precision_recall(item_ground_truth, item_based_recommendations)
print(f"Item-Based Collaborative Filtering - Precision: {item_precision}, Recall: {item_recall}")

# Content-Based Filtering
content_precision, content_recall = calculate_precision_recall(content_ground_truth, content_based_recommendations)
print(f"Content-Based Filtering - Precision: {content_precision}, Recall: {content_recall}")

# Hybrid Filtering
hybrid_precision, hybrid_recall = calculate_precision_recall(
    hybrid_ground_truth,
    lambda user_id: hybrid_recommendations(user_id, "Sunset Boulevard", 100)
)
print(f"Hybrid Filtering - Precision: {hybrid_precision}, Recall: {hybrid_recall}")
```

```python
def compare_recommendations(user_id, movie_title, movie_id, n_recommendations=10):
    """
    Compare recommendations from User-Based, Item-Based, Content-Based, and Hybrid models.
    Includes error handling to ensure no NoneType issues.
    """
    print(f"\nComparing Recommendations for User ID: {user_id}, Movie: {movie_title}\n")

    # User-Based Recommendations
    try:
        print("User-Based Collaborative Filtering:")
        user_recommendations = recommend_movies_user(user_id, n_recommendations)
        if user_recommendations is None:
            print("No recommendations found for User-Based Collaborative Filtering.")
        else:
            for idx, (movie_id, distance) in enumerate(user_recommendations, 1):
                print(f"{idx}: {movie_title_lookup.get(movie_id, 'Unknown')} --- Distance: {distance:.4f}")
    except Exception as e:
        print(f"Error in User-Based Collaborative Filtering: {e}")

    print("\nItem-Based Collaborative Filtering:")
    # Item-Based Recommendations
    try:
        item_recommendations = recommend_movies_item(movie_id, n_recommendations)
        if item_recommendations is None:
            print("No recommendations found for Item-Based Collaborative Filtering.")
        else:
            for idx, (similar_movie_id, distance) in enumerate(item_recommendations, 1):
                print(f"{idx}: {movie_title_lookup.get(similar_movie_id, 'Unknown')} --- Distance: {distance:.4f}")
    except Exception as e:
        print(f"Error in Item-Based Collaborative Filtering: {e}")

    print("\nContent-Based Filtering:")
    # Content-Based Recommendations
    try:
        content_recommendations = get_content_based_recommendations(movie_title, n_recommendations)
        if content_recommendations is None:
            print("No recommendations found for Content-Based Filtering.")
        else:
            for idx, (movie, score) in enumerate(content_recommendations, 1):
                print(f"{idx}: {movie} --- Similarity Score: {score:.4f}")
    except Exception as e:
        print(f"Error in Content-Based Filtering: {e}")

    print("\nHybrid Model Recommendations:")
    # Hybrid Recommendations
    try:
        hybrid_recommendations_list = hybrid_recommendations(user_id, movie_title, n_recommendations)
        if hybrid_recommendations_list is None:
            print("No recommendations found for Hybrid Model.")
        else:
            for idx, (movie, score) in enumerate(hybrid_recommendations_list, 1):
                print(f"{idx}: {movie} --- Hybrid Score: {score:.4f}")
    except Exception as e:
        print(f"Error in Hybrid Recommendations: {e}")

# Example Usage
user_id = 1  # Replace with a valid user ID
movie_title = "Sunset Boulevard"  # Replace with a valid movie title
movie_id = 31  # Add Movie ID explicitly
compare_recommendations(user_id, movie_title, movie_id)
```

```python
def content_hybrid_model_comparison(user_id, movie_title, n_recommendations=10):
    """
    Compare Content-Based and Hybrid models.
    """
    # Store recommendations and scores
    recommendations = {
        "Content-Based": [],
        "Hybrid": []
    }

    # Content-Based Recommendations
    try:
        content_recommendations = get_content_based_recommendations(movie_title, n_recommendations)
        if content_recommendations:
            recommendations["Content-Based"] = [(rec[0], rec[1]) for rec in content_recommendations]
    except Exception as e:
        print(f"Error in Content-Based Filtering: {e}")

    # Hybrid Recommendations
    try:
        hybrid_recommendations_list = hybrid_recommendations(user_id, movie_title, n_recommendations)
        if hybrid_recommendations_list:
            recommendations["Hybrid"] = [(rec[0], rec[1]) for rec in hybrid_recommendations_list]
    except Exception as e:
        print(f"Error in Hybrid Recommendations: {e}")

    # 1. Consistency of Scores
    consistency_data = {}
    for model, recs in recommendations.items():
        if recs:
            scores = [score for _, score in recs if score is not None]
            consistency_data[model] = {
                "Mean Score": np.mean(scores) if scores else None,
                "Score Variance": np.var(scores) if scores else None,
                "Coverage": len(recs)
            }

    # Print Consistency Data
    print("\nConsistency of Scores and Coverage (Content and Hybrid Models):")
    for model, data in consistency_data.items():
        print(f"{model}: Mean = {data['Mean Score']:.4f}, Variance = {data['Score Variance']:.4f}, Coverage = {data['Coverage']}")

    # Visualization of Scores
    plt.figure(figsize=(12, 6))
    for model, recs in recommendations.items():
        if recs:
            scores = [score for _, score in recs]
            plt.plot(range(1, len(scores) + 1), scores, marker='o', label=model)

    plt.title("Content-Based and Hybrid Models: Consistency of Scores", fontsize=16)
    plt.xlabel("Rank", fontsize=14)
    plt.ylabel("Distance/Score", fontsize=14)
    plt.legend()
    plt.grid()
    plt.tight_layout()
    plt.show()

    # Conclude the Best Model
    best_model = max(consistency_data, key=lambda x: (consistency_data[x]['Coverage'], -consistency_data[x]['Score Variance']))
    print(f"\n**Conclusion:** The best model between Content-Based and Hybrid is **{best_model}**.")

# Example Usage
user_id = 1
movie_title = "Sunset Boulevard"
content_hybrid_model_comparison(user_id, movie_title)
```
